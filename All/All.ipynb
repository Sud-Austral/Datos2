```python
import pandas as pd

data_paises = pd.read_csv("../InformacionGeneral/Internacional/Detalle_paises.csv")
del data_paises["WORLD_SN"]
del data_paises["ECDC_SN"]
del data_paises["OURWOR_SN"]
del data_paises["JHONH_SN"]
del data_paises["Early_SN"]

data_woldmeter = pd.read_csv("../worldometers.info/worldometers.csv")
columna_woldmeter = data_woldmeter.columns[1:]

for i in columna_woldmeter:
    data_woldmeter[i+"Worldmeter"] = data_woldmeter[i]
    del data_woldmeter[i]
    
merged_left = pd.merge(left=data_paises, right=data_woldmeter, how='left', left_on='WORLDM_Country,Other', right_on='Country,Other')

def DataframeTotales():
    ds = pd.read_csv("../ecdc.europa/ecdc_europa.csv")
    valores = {'Countries and territories': ds.countriesAndTerritories,
        'cases': ds.cases, 'deaths' : ds.deaths
        }
    df = pd.DataFrame(valores)
    res = df.groupby(['Countries and territories']).sum()
    #print(res)
    #res.to_csv("ecdc_totales.csv", index = False)
    return res
data_aux = DataframeTotales()
data_aux["casesECDC"] = data_aux["cases"]
del data_aux["cases"]
data_aux["deathsECDC"] = data_aux["deaths"]

merged_left = pd.merge(left=data_paises, right=data_aux, how='left', left_on='ECDC_countriesAndTerritories', right_on='Countries and territories')


#data_europa = pd.read_csv("../ecdc.europa/ecdc_europa.csv")
#data_europa
data_ourworld_con = pd.read_csv("../ourworldindata.org/total-cases-covid-19.csv")
data_ourworld_dea = pd.read_csv("../ourworldindata.org/total-deaths-covid-19.csv")
#data_ourworld["Entity"].unique()
#data_ourworld
paises = []
confirmados = []
muertes = []
for i in data_ourworld_con["Entity"].unique():
    maximoConfirmado = max(data_ourworld_con[data_ourworld_con["Entity"] == i]["Total confirmed cases of COVID-19 (cases)"])
    maximoMuerte = max(data_ourworld_dea[data_ourworld_dea["Entity"] == i]["Total confirmed deaths due to COVID-19 (deaths)"])
    #print(i + " " + str(maximo))
    paises.append(i)
    confirmados.append(maximoConfirmado)
    muertes.append(maximoMuerte)

data_ourworld = pd.DataFrame({
    "paises": paises,
    "confirmadoOURWORLD":confirmados,
    "muertosOURWORLD": muertes    
}
)
merged_left = pd.merge(left=merged_left, right=data_ourworld, how='left', left_on='OURWORLD_Entity', right_on='paises')
del merged_left["paises"] 


data_jhon = pd.read_csv("../Johns_Hopkins-covid19/diario/ultimoRegistro.csv")
data_jhon

paises = []
confirmados = []
muertes = []
recuperados = []
activos = []
for i in data_jhon["Country_Region"].unique():
    maximoConfirmado = sum(data_jhon[data_jhon["Country_Region"] == i]["Confirmed"])
    maximoMuerte = sum(data_jhon[data_jhon["Country_Region"] == i]["Deaths"])
    maximoRecuperado = sum(data_jhon[data_jhon["Country_Region"] == i]["Recovered"])
    maximoActivo = sum(data_jhon[data_jhon["Country_Region"] == i]["Active"])
    #print(i + " " + str(maximo))
    paises.append(i)
    confirmados.append(maximoConfirmado)
    muertes.append(maximoMuerte)
    recuperados.append(maximoRecuperado)
    activos.append(maximoActivo)
    
data_jhon_salida = pd.DataFrame({
    "paises": paises,
    "confirmadoJHON":confirmados,
    "muertosJHON": muertes,
    "recuperadosJHON": recuperados,
    "activosJHON": activos
}
)
merged_left = pd.merge(left=merged_left, right=data_jhon_salida, how='left', left_on='JHONH_Country/Region', right_on='paises')
del merged_left["paises"] 

data_early = pd.read_csv("../EarlyAlert/Current_Coronavirus_Cases_and_Deaths.csv")
#data_early.to_excel("test.xlsx")
data_early = data_early.fillna(0)

paises = []
confirmados = []
muertes = []
recuperados = []
for i in data_early["COUNTRY"].unique():
    maximoConfirmado = sum(data_early[data_early["COUNTRY"] == i]["Confirmed"])
    maximoMuerte = sum(data_early[data_early["COUNTRY"] == i]["Deaths"])
    #print(i + " " + str(maximo))
    paises.append(i)
    confirmados.append(maximoConfirmado)
    muertes.append(maximoMuerte)
    
    
data_early_salida = pd.DataFrame({
    "paises": paises,
    "confirmadoEarly":confirmados,
    "muertosEarly": muertes
}
)

merged_left = pd.merge(left=merged_left, right=data_early_salida, how='left', left_on='Early_COUNTRY', right_on='paises')
del merged_left["paises"] 



merged_left.to_csv("All.csv", index=False)
merged_left.to_excel("All.xlsx", index=False)
```
